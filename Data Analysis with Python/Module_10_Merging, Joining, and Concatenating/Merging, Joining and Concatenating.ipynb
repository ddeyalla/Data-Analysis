# The pd.concat() Method, Part 1

pandas.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)[source]¶
Concatenate pandas objects along a particular axis with optional set logic along the other axes.

Can also add a layer of hierarchical indexing on the concatenation axis, which may be useful if the labels are the same (or overlapping) on the passed axis number.

```python
pd.concat([week_sales_1, week_sales_2], ignore_index = True)
# ignore index to set a new index values based on the no of items
#the above dataframes have identical columns data and index
```

# The pd.concat() Method, Part 2

# Inner Joins

```python
week_sales_2.head()
```

```python
pd.merge(week_sales_1, week_sales_2, how = "inner", on = "Customer ID", suffixes=[" Week 1", " Week 2"]).head()
```

# Outer Joins

```python
merged = pd.merge(week_sales_1, week_sales_2, how = "outer", on = ["Customer ID"] , indicator = True) # indicator indicates from where the data is being pulled from
```

```python
mask = merged["_merge"].isin(["both"]) #left_only, right_only
merged[mask]
```

# Left Joins

```python
left = pd.merge(week_sales_1, foods, how = "left", indicator = True)
```

```python
left.head()
```

```python
mask = left["_merge"].isin(["both"]) #left_only, both by default
left[mask]
#left only
```

# The left_on and right_on Parameters

```python
week_sales_2.head()
```

```python
customers.head()
```

```python
pd.merge(week_sales_2, customers, how = "left", left_on = "Customer ID", right_on = "ID", sort = True).drop("ID", axis = "columns")
```

# Merging by Indexes with the left_index and right_index Parameters

```python
customers.head()
```

```python
foods.head()
```

```python
sales = pd.merge(week_sales_1, customers, how = "left", left_on="Customer ID", right_index= True)
pd.merge(sales, foods, how = "left", left_on= "Food ID", right_index= True)
```
