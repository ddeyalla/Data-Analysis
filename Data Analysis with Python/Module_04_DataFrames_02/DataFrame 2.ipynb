# BOOLEN OPERATIONS

## Filter with More than One Condition (AND - &)



```python
employee.head()
```

```python
Male_mask = employee["Gender"] == "Male"
```

```python
Marketing_mask = employee["Team"] == "Marketing"
```

```python
employee[Male_mask | Marketing_mask].head() # & / |
```

```python
mask_1 = employee["First Name"] == "Roberts"
```

```python
mask_2 = employee["Team"] == "Client Services"
```

```python
mask_3 = employee["Start Date"] > "2016-06-01"
```

```python
employee[(mask_1 & mask_2) | mask_3]
```
