```python
import pandas as pd
from pandas_datareader import data as dt
import matplotlib.pyplot as plt
```

```python
goog = dt.DataReader(name = "GOOG", data_source = "yahoo", start= "2016-01-01", end = "2019-01-10")
```

# Bar Graphs

```python
goog.plot.barh("High", "Low", figsize=(20,10))
```

```python
def performance(stock_price):
    if stock_price <= 900:
        return "Poor"
    elif stock_price > 900 and stock_price <= 1000:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
plt.style.use("ggplot")
goog["High"].apply(performance).value_counts().plot.bar(figsize = (10,10))
```

```python
plt.style.use("ggplot")
goog["High"].apply(performance).value_counts().plot.pie(figsize = (10,10))
```

```python
plt.style.use("ggplot")
goog["High"].apply(performance).value_counts().plot.hist(figsize = (10,10))
```
